```python
import pathpy as pp
import sqlite3
```

```python
con = sqlite3.connect('../../data/temporal_networks.db')
con.row_factory = sqlite3.Row
tempnet = pp.TemporalNetwork.fromSQLite(con.execute('SELECT source, target, time FROM ants_1_1'), directed=False)
```

```python
paths = pp.PathExtraction.TemporalPaths.extract(tempnet, delta=10)
network = pp.HigherOrderNetwork(paths)
```
