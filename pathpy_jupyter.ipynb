```python
import pathpy as pp
import sqlite3
```

```python
con = sqlite3.connect('../../data/temporal_networks.db')
con.row_factory = sqlite3.Row
t = pp.TemporalNetwork.fromSQLite(con.execute('SELECT source, target, time FROM sociopatterns_workplace'), directed=False)
```

```python
paths = pp.PathExtraction.TemporalPaths.extract(t, delta=10)
network = pp.HigherOrderNetwork(paths)
```
